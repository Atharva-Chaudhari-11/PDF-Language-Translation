```python
bounds = reader.readtext(np.array(images[0]), min_size=0, slope_ths=0.2,ycenter_ths=0.7,height_ths=0.6,width_ths=0.8,decoder='beamsearch',beamWidth=10)
bounds
```

```python
def draw_boxes(image,bound,color='yellow',width=2):
    draw= ImageDraw.Draw(image)
    for bound in bounds:
        p0,p1,p2,p3 = bound[0]
        draw.line([*p0,*p1,*p2,*p3,*p0], fill=color, width=width)
    return image
draw_boxes(images[0],bounds)
```

```python
# this is some of the module or libraries are required
import os
import sys
import transformers
import tensorflow as tf
from datasets import load_dataset
from transformers import AutoTokenizer
from transformers import TFAutoModelForSeq2SeqLM, DataCollatorForSeq2Seq
from transformers import AdamWeightDecay
from transformers import AutoTokenizer, TFAutoModelForSeq2SeqLM
```

```python
# loadting the model helsinki from huggingface
model_checkpoint = "Helsinki-NLP/opus-mt-en-hi"
```

```python
# Helsinki-NLP/opus-mt-en-hi model
# source: https://huggingface.co/Helsinki-NLP/opus-mt-en-hi
```

```python
# The Dataset
# Source: https://huggingface.co/datasets/cfilt/iitb-english-hindi
```

```python
raw_datasets = load_dataset("cfilt/iitb-english-hindi")
```

```python
raw_datasets
```

```python
raw_datasets['train'][1]
```

```python
# Preprocessing the data
```

```python
# <sglya words sathi> it is important to tokenized the each words into the index number mhnun used.
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)
```

```python
# for sample test to check the tokenizer works or not
tokenizer("Hello, our project name is textBridge!")
```

```python
# this is the targated text
with tokenizer.as_target_tokenizer():
    print(tokenizer(["एक्सेर्साइसर पहुंचनीयता अन्वेषक"]))
```

```python
max_input_length = 128
max_target_length = 128

source_lang = "en"
target_lang = "hi"


def preprocess_function(examples):
    inputs = [ex[source_lang] for ex in examples["translation"]]
    targets = [ex[target_lang] for ex in examples["translation"]]
    model_inputs = tokenizer(inputs, max_length=max_input_length, truncation=True)

    # Setup the tokenizer for targets
    with tokenizer.as_target_tokenizer():
        labels = tokenizer(targets, max_length=max_target_length, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs
```

```python
preprocess_function(raw_datasets["train"][:2])
```

```python
# it tokenized dataset take the all of data and do the tokenization here...
```

```python
tokenized_datasets = raw_datasets.map(preprocess_function, batched=True)
```

```python
# load the model it is for seq2seq learning
model = TFAutoModelForSeq2SeqLM.from_pretrained(model_checkpoint)
```

```python
batch_size = 16
learning_rate = 2e-5
weight_decay = 0.01
num_train_epochs = 1
```

```python
# data-collector take the data into batches and pass it to model.
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf")
```

```python
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)
```

```python
pip install tensorflow_datasets
```

```python
# now here im load the data for training
train_dataset = model.prepare_tf_dataset(
    tokenized_datasets['train'],
    batch_size=batch_size,
    shuffle=True,
    collate_fn=data_collator,
)
```

```python
validation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=batch_size,
    shuffle=False,
    collate_fn=data_collator,
)
```

```python
generation_dataset = model.prepare_tf_dataset(
    tokenized_datasets["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator,
)
```

```python
optimizer = AdamWeightDecay(learning_rate=learning_rate, weight_decay_rate=weight_decay)
model.compile(optimizer=optimizer)
```

```python
model.fit(train_dataset, validation_data=validation_dataset, epochs=1)
```

```python
# to save the model into the drive
model.save_pretrained("tf_model/")
```
